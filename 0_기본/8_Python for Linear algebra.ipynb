## Numpy를 이용한 선형계 문제

image.png

```python
matrix_a = np.array([[1,2,3],[2,5,3],[1,0,8]])
matrix_a_inverse = np.linalg.inv(matrix_a)
b = np.array([[5],[3],[17]])

x = np.dot(matrix_a_inverse,b)
x
```

```python
matrix_a = np.array([[1,2,3],[2,5,3],[1,0,8]])
b = np.array([[5],[3],[17]])

x = np.linalg.solve(matrix_a, b)
x
```

## Numpy를 이용한 동력계 문제
동력계(dynamic system)는 시간에 따라 값이 변하는 변수들의 유한집합 <br>
어떤 시점(t)에서의 변수의 값을 변수의 상태(state of variable)이라 하고, 생성된 벡터를 동력계의 상태(state of the dynamical system) <br> <br>

Ex) 경쟁관계에 있는 두 개의 텔레비전 채널 1과 2는 초기에 각각 시장 점유율이 50%이다. 1년 동안 채널 1은 채널 2의 시청률을 10% 잠심하고, 채널 2는 채널 1의시청률 20%를 잠식한다고 가정하자. 5년 후 각 채널의 시장점유율은 어떻게 될것인가?

```python
matrix_a = np.array([[0.8, 0.1],
                    [0.2, 0.9]])

init_vector = np.array([0.5, 0.5])
```

```python
x0 = init_vector
print(x0)
```

```python
x1 = matrix_a.dot(x0)
print(x1)
```

```python
x2 = matrix_a.dot(x1)
print(x2)
```

```python
def x(t):
    matix_a = np.array([[0.8, 0.1],
                        [0.2, 0.9]])
    init_vector = np.array([0.5, 0.5])

    if t == 0:
        xt = init_vector
        return xt
    
    else:
        xt = init_vector
        for i in range(t):
            xt = matrix_a.dot(xt)
        return xt
```

```python
print('초기 각 채널 점유율: ', x(0))
print('1년후의 각 채널 점유율: ', x(1))
print('2년후의 각 채널 점유율: ', x(2))
```

```python
print('5년후의 각 채널 점유율: ', x(5))
print('100년후의 각 채널 점유율: ', x(100))
```
